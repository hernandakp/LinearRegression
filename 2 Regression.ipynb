**Regresi**

```python
# Melihat jenis-jenis dari product_category
df.product_category.unique()
```

**1. Regresi Penjualan Produk Berdasarkan Kategori "Intimates"**

```python
# Memplot data berdasarkan kategori "Intimates"
intimates = df[df.product_category.isin(['Intimates'])]
intimates
```

```python
# Encode data
X = intimates['sold_at'].values
Y = intimates['sale_price'].values

# Regresi Metode Least-Square Linear

# Menghitung slope, intercept, dan nilai rata-rata
X_avg = np.average(X)
Y_avg = np.average(Y)
X_mean = np.mean(X)
Y_mean = np.mean(Y)
num = 0
den = 0
for i in range(len(X)):
  num += (X[i] -  X_mean)*(Y[i] - Y_mean)
  den += (X[i] - X_mean)**2
m = num/den
c = Y_mean - m*X_mean
print("Nilai yang didapat:")
print("Slope =", (m))
print("Intercept =", (c))
print("X rata-rata =", (X_avg))
print("Y rata-rata =", (Y_avg))

# Menghitung letak Y
Y_pred = m*X + c
print ("Titik Y =", Y_pred)

# Memplot grafik
plt.figure(figsize = (4, 5))
plt.title("Regresi Penjualan Produk Berdasarkan Kategori 'Intimates'")
plt.scatter(X, Y, color = "red")
plt.plot(X, Y_pred, color = "green")
plt.xlabel("sold_at")
plt.ylabel("sale_price")
plt.show()
```

```python
# Mengubah tipe data
col_int = ['user_id']
for col in col_int:
    df[col] = pd.to_numeric(df[col])
for col in ['product_category', 'product_name', 'product_brand', 'product_sku', 'created_at', 'sold_date']:
    df[col] = df[col].astype(np.str_)
df.info()
```

```python
# Product yang terjual dengan pendapatan tertinggi
df_sales = df.groupby('product_name').sum()[['sale_price']]
df_sales.sort_values(by=['sale_price'], ascending=False).head()
```

```python
df_month_year = df.groupby('sold_date').sum()[['sale_price']]
df_month_year = df_month_year.iloc[:-1]
print(df_month_year.sort_values(by=['sale_price'], ascending=False))
sales = df.groupby('sold_date').sum()['sale_price'].round(2)
sales.plot(kind='line', x='Month-Year', y='sale_price', figsize=(12,8))
plt.legend()
plt.grid()
```
